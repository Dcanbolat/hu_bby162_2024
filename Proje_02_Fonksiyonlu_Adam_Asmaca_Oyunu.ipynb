```python
#Proje 02: Fonksiyonlu Adam Asmaca Oyunu

from random import choice

def oyunu_baslat():
    print("Adam Asmaca Oyununa Hoş Geldiniz!")
    print("Lütfen bir zorluk seviyesi seçiniz:")
    print("1. Kolay (7 can)")
    print("2. Orta (5 can)")
    print("3. Zor (3 can)")
    secilen_seviye = input("Seviye seçiniz (1/2/3): ")

    if secilen_seviye == '1':
        adamCan = 7
    elif secilen_seviye == '2':
        adamCan = 5
    elif secilen_seviye == '3':
        adamCan = 3
    else:
        print("Geçersiz seçim. Lütfen tekrar deneyin.")
        oyunu_baslat()
        return

    kelimeler = ["bisiklet", "triatlon", "yüzme", "koşu"]
    secilenKelime = choice(kelimeler)

    dizilenKelime = []
    for diz in secilenKelime:
        dizilenKelime.append("_")
    print("Tahmin edilmesi gereken kelime: ", " ".join(dizilenKelime))

    while adamCan > 0:
        girilenHarf = input("Bir harf giriniz: ").lower()
        if not girilenHarf.isalpha() or len(girilenHarf) != 1:
            print("Geçersiz giriş. Lütfen bir harf giriniz.")
            continue

        if girilenHarf in secilenKelime:
            for i in range(len(secilenKelime)):
                if secilenKelime[i] == girilenHarf:
                    dizilenKelime[i] = girilenHarf
            print("Tahmin edilen kelime: ", " ".join(dizilenKelime))
            if "_" not in dizilenKelime:
                print("Tebrikler, kelimeyi doğru tahmin ettiniz!")
                break
        else:
            adamCan -= 1
            print("Bilemediniz! Kalan can: ", adamCan)
            if adamCan == 0:
                print("Malesef, tüm hakkınızı kullandınız. Doğru kelime: ", secilenKelime)
                break

    tekrar_oyna = input("Tekrar oynamak istiyor musunuz? (Evet/Hayır): ").lower()
    if tekrar_oyna == "evet":
        oyunu_baslat()
    else:
        print("Oyundan çıkılıyor. İyi günler!")
        return

oyunu_baslat()
```
